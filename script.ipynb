# Import Libraries

mengimport package pandas dan numpy, pandas untuk memmbaca data dan membuat dataframe, numpy unutk pengolahan matrix

```python
import pandas as pd
import numpy as np
```

# Import Data
menggunakan pd.read_excel untuk membaca data excel, parameter sheet_name berguna untuk menentukan sheet yang akan dibaca, index_col berguna untuk menentukan kolom yang akan dijadikan index

```python
df = pd.read_excel("data.xlsx",sheet_name="Sheet2",index_col=0)
df_input = pd.read_excel("data.xlsx",sheet_name="Sheet3")
df_tenaga_kerja = pd.read_excel("data.xlsx",sheet_name="Sheet4")
```

# Menghitung Matriks A
untuk menghitung matriks A, dapat diperoleh dengan membagi nilai dari tiap kolom dengan nilai dari data df_input

```python
df_A = df.copy()
df_A = df_A.div(df_input.iloc[0], axis=1)
df_A
```

untuk mengubah menjadi matriks, kita dapat menggunakan method .values

```python
matrix_A = df_A.values
#set matrix view so it's not truncated
np.set_printoptions(threshold=np.inf)

#set decimal precision
np.set_printoptions(precision=3)

print(matrix_A)
```

# Membuat Matriks Identitas
untuk membuat matriks identitas, kita dapat menggunakan method np.identity(n) dengan n adalah ukuran matriks identitas yang akan dibuat

```python
# buat matriks identitas 17x17
matrix_I = np.identity(17)
print(matrix_I)
```

# menghitung matriks (I-A)
perhitungan ini dapat dilakukan dengan cara mengurangkan matriks I dengan matriks A

```python
matrix_I_A = matrix_I - matrix_A
print(matrix_I_A)
```

# menghitung inverse matriks (I-A)
untuk menghitung inverse matriks, kita menggunakan fungsi np.linalg.inv(m) dengan m adalah matriks yang akan dihitung

```python
matrix_I_A_inv = np.linalg.inv(matrix_I_A)
matrix_I_A_inv
```

# analisis keterkaitan ke belakang (tiap variabel)
perhitungan ini dapat dilakukan dengan menjumlahkan tiap baris dari matriks inverse (I-A). perhitungan dapat dilakukan dengan menggunakan method .sum(axis=0)

```python
matrix_I_A_inv.sum(axis=0)
```

untuk mempermudah pembacaan, dapat diubah menjadi bentuk dataframe dengan kolom yang sesuai

```python
df_belakang = pd.DataFrame(matrix_I_A_inv.sum(axis=0).reshape(1,17),columns=df.columns)
df_belakang
```

menghitung rata-rata dari df_belakang

```python
#rata-rata 
df_belakang.mean(axis=1) # agar dirata-ratakan semua kolom, maka menggunakan axis = 1
```

# analisis keterkaitan ke depan (tiap baris)
perhitungan ini dapat dilakukan dengan menjumlahkan tiap kolom dari matriks inverse (I-A). perhitungan dapat dilakukan dengan menggunakan method .sum(axis=1)

```python
matrix_I_A_inv.sum(axis=1)
```

untuk mempermudah pembacaan, dapat diubah menjadi bentuk dataframe dengan kolom yang sesuai

```python
#analisis keterkaitan ke depan (tiap baris)
df_depan = pd.DataFrame(matrix_I_A_inv.sum(axis=1).reshape(17,1),index=df.index,columns=["keterkaitan ke depan"])
df_depan
```

# Menghitung IDP
perhitungan IDP dapat dilakukan dengan membagi nilai dari df_belakang dengan rata-rata dari df_belakang

perhitungan rata-rata dari df_belakng dapat dilakukan dengan .mean(axis = 1), axis 1 diguanakn karena ingin dihitung rata-rata dari semua kolom

```python
df_belakang.mean(axis=1)
```

untuk membagi keseluruhan nilai dengan nilai rata-rata dapat dilakukan dengan .div(a) dengan a adalah pembagi

```python
df_belakang.div(df_belakang.mean(axis=1),axis = 0) #axis = 0 agar dirata-ratakan semua baris
```

# Menghitung IDK
perhitungan IDK dapat dilakukan dengan membagi nilai dari df_depan dengan rata-rata dari df_depan

rata-rata df_depan dapat dihitung dengan cara .mean()

```python
df_depan.mean(axis = 0)
```

untuk membagi keseluruhan nilai dengan nilai rata-rata dapat dilakukan dengan .div(a) dengan a adalah pembagi

```python
#hitung IDK 
df_depan.div(df_depan.mean(axis = 0),axis = 1)
```

# Koefisien Tenaga Kerja

koefisien tenaga kerja dapat dihitung dengan membagi df_tenaga_kerja dengan df_input

```python
koef_tk = df_tenaga_kerja.div(df_input.iloc[0],axis = 1) #iloc[0] menunjukkan baris pertama dari df_input
koef_tk
```

# Angka Pengganda Tenaga Kerja
dapat dihitung melalui perkalian matriks koef_tk dengan inverse matrix(I-A)

membuat matriks koef_tk dapat dilakukan dengan .values

```python
matriks_tk = koef_tk.values
matriks_tk
```

perkalian matriks dapat dilakukan dengan fungsi dot

```python
np.dot(matriks_tk,matrix_I_A_inv)
```

agar lebih rapi, dapat diubah menjadi dataframe

```python
pengganda_tk = pd.DataFrame(np.dot(matriks_tk,matrix_I_A_inv),columns=df_tenaga_kerja.columns)
pengganda_tk
```

hitung pengganda 100 dan 1000

```python
pengganda_tk.multiply(100)
```

```python
pengganda_tk.multiply(1000)
```
